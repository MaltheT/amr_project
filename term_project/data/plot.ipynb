```python
import matplotlib.pyplot as plt
import numpy as np
import pylab

plt.rcParams['figure.figsize'] = [10, 5]
```

```python
"""
Hovering: z data 
"""
z_data_1 = np.genfromtxt('z_050001.csv', delimiter=',')
z_data_2 = np.genfromtxt('z_500130.csv', delimiter=',')
z_data_3 = np.genfromtxt('z_2500110.csv', delimiter=',')
z_data_4 = np.genfromtxt('z_30007520.csv', delimiter=',')

#simulation time
simtime = 30

#plot trajectory of drone height from initial position to target pose
t_1 = np.linspace(0, simtime, len(z_data_1))
t_2 = np.linspace(0, simtime, len(z_data_2))
t_3 = np.linspace(0, simtime, len(z_data_3))
t_4 = np.linspace(0, simtime, len(z_data_4))


#t_r = np.linspace(0, simtime, len(distance_1))
#reference = np.zeros(len(t_r))

plt.plot(t_1, z_data_1, color='blue', label='Kp = 0.5, Ki = 0.0, Kd = 0.1')
plt.plot(t_2, z_data_2, color='orange', label='Kp = 5.0, Ki = 0.1, Kd = 3.0')
plt.plot(t_3, z_data_3, color='green', label='Kp = 2.5, Ki = 0.01, Kd = 1.0')
plt.plot(t_4, z_data_4, color='red', label='Kp = 3.0, Ki = 0.075, Kd = 2.0')

plt.xlabel('Time/s')
plt.ylabel('Height/m')
plt.legend()
```

```python
"""
Pathplanning
"""
#simulation data
x_data_m = np.genfromtxt('x.csv', delimiter=',')
y_data_m = np.genfromtxt('y.csv', delimiter=',')
z_data_m = np.genfromtxt('z.csv', delimiter=',')
distance_m = np.genfromtxt('distance.csv', delimiter=',')
yaw_m = np.genfromtxt('yaw.csv', delimiter=',')
desired_angle_m = np.genfromtxt('desired_angle.csv', delimiter=',')

x_data_sm = np.genfromtxt('x_smooth.csv', delimiter=',')
y_data_sm = np.genfromtxt('y_smooth.csv', delimiter=',')
z_data_sm = np.genfromtxt('z_smooth.csv', delimiter=',')
distance_sm = np.genfromtxt('distance_smooth.csv', delimiter=',')
yaw_sm = np.genfromtxt('yaw_smooth.csv', delimiter=',')
desired_angle_sm = np.genfromtxt('desired_angle_smooth.csv', delimiter=',')

y_data_sm = y_data_sm[:-1]

#simulation time
simtime = 180 


"""
Plotting x data 
"""
t = np.linspace(0, simtime, len(x_data_m))

plt.plot(t, x_data_m, color='blue', label='x')
plt.xlabel('Time/s')
plt.ylabel('x')
plt.legend()
```

```python
"""
Plotting y data
"""
t = np.linspace(0, simtime, len(y_data_m))

plt.plot(t, y_data_m, color='red', label='y')
plt.xlabel('Time/s')
plt.ylabel('y')
plt.legend()
```

```python
"""
Plotting z data
"""
t = np.linspace(0, simtime, len(z_data_m))

plt.plot(t, z_data_m, color='green', label='z')
plt.xlabel('Time/s')
plt.ylabel('z')
plt.legend()
```

```python
"""
Desired distance
"""
t = np.linspace(0, simtime, len(distance_m))

plt.plot(t, distance_m, color='purple', label='Error between current and desired distance')
plt.xlabel('Time/s')
plt.ylabel('Distance Error')
plt.legend()
```

```python
"""
3D trajectory of the path 
"""
#waypoints 
waypoints_x = [-3, -3, 3, 3, 0]
waypoints_y = [0, 4.5, 4, -4, 0]
waypoints_z = [2.5, 2.5, 2, 2, 0]

#plotting the figure
fig = plt.figure()
ax = plt.axes(projection='3d')

print(np.shape(x_data_m), np.shape(y_data_m), np.shape(z_data_m))    
pylab.rcParams['figure.figsize'] = (50, 25)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter3D(waypoints_x, waypoints_y, waypoints_z, color='red')
ax.scatter3D(x_data_m, y_data_m, z_data_m, color='darkgreen', alpha=0.1)

```

```python
"""
2D trajectory of the path 
"""
pylab.rcParams['figure.figsize'] = (25, 15)

#plotting the figure
plt.plot(x_data_m, y_data_m, color='darkgreen', label='2D Trajectory')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(waypoints_x, waypoints_y, color='red')
plt.legend()
```

```python
"""
Difference between yaw angle and desired angle 
"""

t_yaw = np.linspace(0, simtime, len(yaw_m))
t_ref = np.linspace(0, simtime, len(desired_angle_m))

pylab.rcParams['figure.figsize'] = (15, 10)

plt.plot(t_yaw, yaw_m, color='red', label='Yaw angle')
plt.plot(t_ref, desired_angle_m, linestyle='dashed', color='black', label='Desired Yaw Angle')
plt.xlabel('Time/s')
plt.ylabel('Yaw angle/rad')
plt.legend()
```
