```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import pylab
from mpl_toolkits.axes_grid1 import Divider, Size
from mpl_toolkits import mplot3d


plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams.update({'font.size': 18})
```

```python
"""
Hovering: z data 
"""
z_data_1 = np.genfromtxt('z_test1.csv', delimiter=',')
z_data_2 = np.genfromtxt('z_test2.csv', delimiter=',')
z_data_3 = np.genfromtxt('z_test3.csv', delimiter=',')
z_data_4 = np.genfromtxt('z_test4.csv', delimiter=',')

#simulation time
simtime = 30

#plot trajectory of drone height from initial position to target pose
t_1 = np.linspace(0, simtime, len(z_data_1))
t_2 = np.linspace(0, simtime, len(z_data_2))
t_3 = np.linspace(0, simtime, len(z_data_3))
t_4 = np.linspace(0, simtime, len(z_data_4))


#t_r = np.linspace(0, simtime, len(distance_1))
#reference = np.zeros(len(t_r))

plt.plot(t_1, z_data_1, color='blue', label='Kp = 0.5, Ki = 0.0, Kd = 0.1')
plt.plot(t_2, z_data_2, color='orange', label='Kp = 5.0, Ki = 0.1, Kd = 3.0')
plt.plot(t_3, z_data_3, color='green', label='Kp = 2.5, Ki = 0.01, Kd = 1.0')
plt.plot(t_4, z_data_4, color='red', label='Kp = 3.0, Ki = 0.075, Kd = 2.0')

plt.xlabel('Time/s')
plt.ylabel('Height/m')
plt.legend()
```

```python
"""
Pathplanning data
"""
#simulation data
x_data_sm = np.genfromtxt('x_first.csv', delimiter=',')
y_data_sm = np.genfromtxt('y_first.csv', delimiter=',')
z_data_sm = np.genfromtxt('z_first.csv', delimiter=',')
distance_sm = np.genfromtxt('distance_first.csv', delimiter=',')
yaw_sm = np.genfromtxt('yaw_first.csv', delimiter=',')
desired_angle_sm = np.genfromtxt('desired_angle_first.csv', delimiter=',')

x_data = np.genfromtxt('x_second.csv', delimiter=',')
y_data = np.genfromtxt('y_second.csv', delimiter=',')
z_data = np.genfromtxt('z_second.csv', delimiter=',')
distance = np.genfromtxt('distance_second.csv', delimiter=',')
yaw = np.genfromtxt('yaw_second.csv', delimiter=',')
desired_angle = np.genfromtxt('desired_angle_second.csv', delimiter=',')


#simulation time
simtime = 180 


"""
Plotting x-data 
"""
t = np.linspace(0, simtime, len(x_data_sm))

"""
Plotting x data 
"""
plt.plot(t, x_data_sm, color='red')
plt.xlabel('Time/s')
plt.ylabel('x')
plt.legend()
```

```python
"""
Plotting y-data 
"""
t = np.linspace(0, simtime, len(y_data_sm))

plt.plot(t, y_data_sm, color='green')
plt.xlabel('Time/s')
plt.ylabel('y')
plt.legend()
```

```python
"""
Plotting z-data 
"""
t_1 = np.linspace(0, simtime, len(z_data_sm))
t_2 = np.linspace(0, simtime, len(z_data))

plt.plot(t_1, z_data_sm, color='blue', label='First simulation')
plt.plot(t_2, z_data, color='purple', label='Second simulation')
plt.xlabel('Time/s')
plt.ylabel('z')
plt.legend()
```

```python
"""
Desired distance
"""
t = np.linspace(0, simtime, len(distance_sm))

plt.plot(t, distance_sm, color='purple', label='Error between current and desired distance')
plt.xlabel('Time/s')
plt.ylabel('Distance Error')
plt.legend()
```

```python
"""
3D trajectory of the path 
"""
#waypoints 
waypoints_x = [-3, -3, 3, 3, 0]
waypoints_y = [0, 4.5, 4, -4, 0]
waypoints_z = [2.5, 2.5, 2, 2, 0]

#plotting the figure
fig = plt.figure()
ax = plt.axes(projection='3d')

pylab.rcParams['figure.figsize'] = (50, 25)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter3D(waypoints_x, waypoints_y, waypoints_z, color='red')
ax.scatter3D(x_data_sm, y_data_sm, z_data_sm, color='darkgreen')


```

```python
"""
2D trajectory of the path 
"""
pylab.rcParams['figure.figsize'] = (15, 10)

#plotting the figure
plt.plot(x_data_sm, y_data_sm, color='darkgreen', linestyle='dashed', label='First simulation')
plt.plot(x_data, y_data, color='blue',linestyle='dashed', label='Second simulation')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(waypoints_x, waypoints_y, color='red')
plt.grid()
plt.legend()
```

```python
"""
Difference between yaw angle and desired angle 
"""

t_yaw = np.linspace(0, simtime, len(yaw_sm))
t_ref = np.linspace(0, simtime, len(desired_angle_sm))

pylab.rcParams['figure.figsize'] = (15, 10)

plt.plot(t_yaw, yaw_sm, color='red', label='Yaw angle')
plt.plot(t_ref, desired_angle_sm, linestyle='dashed', color='black', label='Desired Yaw Angle')
plt.xlabel('Time/s')
plt.ylabel('Yaw angle/rad')
plt.legend()
```
