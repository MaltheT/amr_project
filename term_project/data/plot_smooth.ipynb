```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import pylab
from mpl_toolkits.axes_grid1 import Divider, Size
from mpl_toolkits import mplot3d


plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams.update({'font.size': 18})
```

```python
"""
Pathplanning data
"""
#simulation data
x_data_sm = np.genfromtxt('x_sm.csv', delimiter=',')
y_data_sm = np.genfromtxt('y_sm.csv', delimiter=',')
z_data_sm = np.genfromtxt('z_sm.csv', delimiter=',')
distance_sm = np.genfromtxt('distance_sm.csv', delimiter=',')
yaw_sm = np.genfromtxt('yaw_sm.csv', delimiter=',')
desired_angle_sm = np.genfromtxt('desired_angle_sm.csv', delimiter=',')

x_data = np.genfromtxt('x.csv', delimiter=',')
y_data = np.genfromtxt('y.csv', delimiter=',')
z_data = np.genfromtxt('z.csv', delimiter=',')
distance = np.genfromtxt('distance.csv', delimiter=',')
yaw = np.genfromtxt('yaw.csv', delimiter=',')
desired_angle = np.genfromtxt('desired_angle.csv', delimiter=',')


#simulation time
simtime = 180 


"""
Plotting x-data 
"""
t = np.linspace(0, simtime, len(x_data_sm))

"""
Plotting x data 
"""
plt.plot(t, x_data_sm, color='red')
plt.xlabel('Time/s')
plt.ylabel('x')
plt.legend()
```

```python
"""
Plotting y-data 
"""
t = np.linspace(0, simtime, len(y_data_sm))

plt.plot(t, y_data_sm, color='green')
plt.xlabel('Time/s')
plt.ylabel('y')
plt.legend()
```

```python
"""
Plotting z-data 
"""
t_1 = np.linspace(0, simtime, len(z_data_sm))
t_2 = np.linspace(0, simtime, len(z_data))

plt.plot(t_1, z_data_sm, color='blue', label='First simulation')
plt.plot(t_2, z_data, color='purple', label='Second simulation')
plt.xlabel('Time/s')
plt.ylabel('z')
plt.legend()
```

```python
"""
Desired distance
"""
t = np.linspace(0, simtime, len(distance_sm))

plt.plot(t, distance_sm, color='purple', label='Error between current and desired distance')
plt.xlabel('Time/s')
plt.ylabel('Distance Error')
plt.legend()
```

```python
"""
3D trajectory of the path 
"""
#waypoints 
waypoints_x = [-3, -3, 3, 3, 0]
waypoints_y = [0, 4.5, 4, -4, 0]
waypoints_z = [2.5, 2.5, 2, 2, 0]

#plotting the figure
fig = plt.figure()
ax = plt.axes(projection='3d')

pylab.rcParams['figure.figsize'] = (50, 25)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter3D(waypoints_x, waypoints_y, waypoints_z, color='red')
ax.scatter3D(x_data_sm, y_data_sm, z_data_sm, color='darkgreen')


```

```python
"""
2D trajectory of the path 
"""
pylab.rcParams['figure.figsize'] = (15, 10)

#plotting the figure
plt.plot(x_data_sm, y_data_sm, color='darkgreen', linestyle='dashed', label='First simulation')
plt.plot(x_data, y_data, color='blue',linestyle='dashed', label='Second simulation')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(waypoints_x, waypoints_y, color='red')
plt.grid()
plt.legend()
```

```python
"""
Difference between yaw angle and desired angle 
"""

t_yaw = np.linspace(0, simtime, len(yaw_sm))
t_ref = np.linspace(0, simtime, len(desired_angle_sm))

pylab.rcParams['figure.figsize'] = (15, 10)

plt.plot(t_yaw, yaw_sm, color='red', label='Yaw angle')
plt.plot(t_ref, desired_angle_sm, linestyle='dashed', color='black', label='Desired Yaw Angle')
plt.xlabel('Time/s')
plt.ylabel('Yaw angle/rad')
plt.legend()
```
